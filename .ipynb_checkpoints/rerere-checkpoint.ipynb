```python
def separate_text(text, symbol, length):
    parts = text.split(symbol)
    if len(parts) > length:
        return symbol.join(parts[:length]), symbol.join(parts[length:])
    else:
        return text, ''
```

```python
hadith_csv = './hadis/musnad_syafii.csv'
output_csv = './output_perawi/output_file.csv'

data = {'ID': [], 'Kitab': [], 'Sanad': [], 'Matan': [], 'Sanad_Arab': [], 'Matan_Arab': []}
```

```python
with open(hadith_csv, newline='', encoding='utf-8') as csvfile:
    hadith_reader = csv.DictReader(csvfile, delimiter=';', quotechar='"')
    for row in hadith_reader:
        id_value = row['id']
        translate_text = row['terjemah']
        arabic_text = row['arab']

        kitab, _, rest = map(str.strip, translate_text.partition(':'))
        sanad, _, matan = map(str.strip, rest.partition('"'))
        kitabArab, _, restArab = map(str.strip, arabic_text.partition(':'))

        sanad_colon_count = sanad.count(':')
        sanad_arab, matan_arab = separate_text(restArab, ":", sanad_colon_count)
        
        data['ID'].append(id_value)
        data['Kitab'].append(kitab)
        data['Sanad'].append(sanad)
        data['Matan'].append(matan)
        data['Sanad_Arab'].append(sanad_arab)
        data['Matan_Arab'].append(matan_arab)
```

```python
df = pd.DataFrame(data)
    
display(df)
```

```python
exceptions = ['bin', 'binti', 'Nabi', 'Rasul', 'Allah', 'Abu', 'Ummu', 'Ummu', 'Imam', 'Al', 'al', 'ibnu', 'dinar']
```

```python
def extractWords(file_path, exceptions):
    words = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            word = line.strip()
            if word not in exceptions:
                words.append(word)
    return words
```

```python
verbaFile = './kamus/verba.txt'
dasarFile = './kamus/dasar.txt'
nominaFile = './kamus/nomina.txt'
stopFile = './kamus/stop.txt'
additionalFile = './kamus/additional.txt'
```

```python
verbaWords = extractWords(verbaFile, exceptions)
dasarWords = extractWords(dasarFile, exceptions)
nominaWords = extractWords(nominaFile, exceptions)
stopWords = extractWords(stopFile, exceptions)
additionalWords = extractWords(additionalFile, exceptions)
```

```python
globArr = []
```

```python
# Create a new CSV file with the "perawi" column

exj = ['shallallahu \'alaihi', 'sallam,', 'sallam', 'kain', 'radliyallahu', '\'anha', '\'anhu', " ", '\'aku', 'isteri', '\'anhuma', 'seekor', 'diberikan', 'orang', 'terpercaya', 'yang', 'shalat', 'saw', 'allah', 'debu', 'utama']
exv = ['nya', 'kah', 'mem']
exz = ["ku"]
exl = ["di"]

with open(input_csv, 'r', encoding='utf-8') as file:
    csv_reader = csv.DictReader(file, delimiter=';')
    fieldnames = csv_reader.fieldnames + ['perawi']

    with open(output_csv, 'w', encoding='utf-8', newline='') as output_file:
        writer = csv.DictWriter(output_file, fieldnames=fieldnames, delimiter=';')
        writer.writeheader()

        for row in csv_reader:
            sanad_sentence = row['Sanad']
            words = sanad_sentence.split()
            perawi_sentence = ''
            previous_word_matched = False

            for word in words:
                if not word.isdigit() and word.lower() not in exj:
                    wordCheck = word.split("-")
                    if (len(wordCheck) > 1):
                        if (wordCheck[0] == wordCheck[1]):
                            word = wordCheck[0]
                    else:
                        word = word.replace('-', '')
                    word = word.replace(':', '').replace('.', '').replace('(', '').replace(')', '').replace('“', '').replace('”', ' ')
                    wordLower = word.lower().rstrip(',').rstrip('.').rstrip(';').rstrip(':').rstrip(')').rstrip('(').rstrip('”').rstrip('“').rstrip(' ').rstrip('\'').rstrip('\"').rstrip('-')
                    if wordLower in extracted_words:
                        if not previous_word_matched:
                            if perawi_sentence == '':
                                perawi_sentence += word
                            else:
                                perawi_sentence += ','
                            previous_word_matched = True
                    else:
                        perawi_sentence += ' ' + word
                        previous_word_matched = False

            # Remove double commas
            perawi_sentence = re.sub(r',+', ',', perawi_sentence).rstrip(',').rstrip('.')

            perawi_words = perawi_sentence.split(", ")
            perawi_words = [phrase.replace(' ' + next((value for value in exj if phrase.endswith(value)), ' '), ' ').strip() for phrase in perawi_words]
            perawi_words = [word for word in perawi_words if word != '']
            
            perawi_words = [word for word in perawi_words if word.lower() not in wordsNewExtra and word.lower() not in exj and word.lower()[-3:] not in exv and word.lower()[-2:] not in exz and word.lower()[:2] not in exl and word.lower()[:3] not in exv]
            globArr.extend(perawi_words)
            perawi_sentence = ', '.join(perawi_words)

            row['perawi'] = perawi_sentence.strip()
            writer.writerow(row)
            print(row['perawi'])

print('New CSV file has been created: new_output_file.csv.')
```

```python
topPerawi = []

from collections import Counter

# Create a Counter object to count the occurrences of each value
value_counts = Counter(globArr)

# Get the most common values and their counts
most_common_values = value_counts.most_common()

# Print the most frequent values
for value, count in most_common_values:
    print(f"Value: {value}, Count: {count}")
    topPerawi.append(value)

# Get the top value
top_value = most_common_values[0][0]
topPerawi.append(top_value)
topPerawi = topPerawi[:5]
print(topPerawi)
```

```python
import csv

# Create an empty array to store the objects
data = []

# Assuming your array is named 'topPerawi'
for element in topPerawi:
    # Create a dictionary object for each element
    obj = {'ID': [], 'Kitab': [], 'Matan': []}
    
    # Add the object to the array
    
    with open('./output_perawi/new_output_file.csv', newline='', encoding='utf-8') as csvfile:
        hadith_reader = csv.DictReader(csvfile, delimiter=';', quotechar='"')
        for row in hadith_reader:
            id_value = row['ID']
            kitab_value = row['Kitab']
            matan_value = row['Matan']
            perawi_value = row['perawi'].split(", ")
            
            if element in perawi_value and matan_value != '' and not matan_value.isdigit():
                obj['ID'].append(id_value)
                obj['Kitab'].append(kitab_value)
                obj['Matan'].append(matan_value)

            
    data.append(obj)
    dfNew = pd.DataFrame(obj)
    print(element)
    display(dfNew)
    directory = "./output_perawi/specific/{}.csv".format(element.replace(" ", "_"))
    dfNew.to_csv(directory, index=False, sep=';')
```

```python
# Install library Gensim and NLTK
# pip install gensim nltk

worExc = ["berkata", "bersabda", "menjawab", "sesungguhnya", "barangsiapa", "al", "si"]

# Import libraries
import numpy as np
import gensim
from gensim import corpora
from pprint import pprint
from nltk.corpus import stopwords

# Download the Indonesian stop words
import nltk
nltk.download('stopwords')

# Get the Indonesian stop words
stop_words = stopwords.words('indonesian')

for element in topPerawi:
    print('\n')
    print('\n')
    print(element)
    print('\n')
    documents = []
    directory = "./output_perawi/specific/{}.csv".format(element.replace(" ", "_"))
    print(directory)
    outputDir = "./output_perawi/lda/output_{}.csv".format(element.replace(" ", "_"))
    
    with open(directory, newline='', encoding='utf-8') as csvfile:
        hadith_reader = csv.DictReader(csvfile, delimiter=';', quotechar='"')
        for row in hadith_reader:
            matan_value = row['Matan'].split(" ")
            matanNew = [x for x in matan_value if x.lower().rstrip(",").strip('"') not in worExc and x.lower().rstrip(",").strip('"') not in stopWords and not any(x.lower().rstrip(",").strip('"') in item.lower() for item in globArr)]
            matanUse = ' '.join(matanNew)
            documents.append(matanUse)
    
    # Tokenize the text
    tokenized_documents = [doc.split() for doc in documents]

    # Remove stop words
    tokenized_documents = [[word for word in doc if word.lower() not in stop_words] for doc in tokenized_documents]

    # Build the dictionary from tokens
    dictionary = corpora.Dictionary(tokenized_documents)

    # Create the corpus (bag-of-words representation)
    corpus = [dictionary.doc2bow(doc) for doc in tokenized_documents]

    # Train the LDA model
    lda_model = gensim.models.LdaModel(corpus, num_topics=5, id2word=dictionary, passes=15)

    # Display topics and keywords
    topics = lda_model.show_topics(num_topics=5, num_words=40)  # Change the number of words to display
    for topic in topics:
        print("Topic {}: {}".format(topic[0], topic[1]))
        print('\n')

    # # Use the LDA model to classify a new document
    # new_document = "Artificial intelligence and machine learning are revolutionizing industries"
    # new_doc_bow = dictionary.doc2bow(new_document.split())
    # topics = lda_model[new_doc_bow]

    # print("Topik Dokumen Baru:")
    # pprint(topics)
```
